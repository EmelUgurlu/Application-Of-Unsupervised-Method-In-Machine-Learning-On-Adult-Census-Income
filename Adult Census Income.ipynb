ALGORITHMS

KMEANS 

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import metrics
range_n_clusters = [2, 3, 4, 5]
silhouette_avg = []
for num_clusters in range_n_clusters:
     
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    silhouette_avg.append(metrics.silhouette_score(X, cluster_labels))   
plt.plot(range_n_clusters,silhouette_avg, 'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Silhouette Score') 
plt.title('Silhouette Analysis for Optimal K')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3, init= 'k-means++', random_state=0)
predict=kmeans.fit_predict(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
df['Cluster']=predict
df.head(10)
```

```python
X=np.array(X)
plt.scatter(X[predict == 0, 0], X[predict == 0, 1], s = 10, c = 'red', label = 'Cluster 1')
plt.scatter(X[predict == 1, 0], X[predict == 1, 1], s = 10, c = 'blue', label = 'Cluster 2')
plt.scatter(X[predict == 2, 0], X[predict == 2, 1], s = 10, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Cluster Centers')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print("V-measure: %0.6f" % metrics.v_measure_score(y,kmeans.labels_))
print("Adjusted Rand Index: %0.6f"
      % metrics.adjusted_rand_score(y, kmeans.labels_))
print("Adjusted Mutual Information: %0.6f"
      % metrics.adjusted_mutual_info_score(y, kmeans.labels_))
print("Silhouette Coefficient: %0.6f"
      % metrics.silhouette_score(X, kmeans.labels_ ))
print("Davies Bouldin Index: %0.6f"
      % metrics.davies_bouldin_score(X, kmeans.labels_ ))
print("Calinski Harabasz Index: %0.6f"
      % metrics.calinski_harabasz_score(X, kmeans.labels_))
```

Optimum number of clusters

```python
from sklearn.cluster import Birch
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
```

```python
sum_of_squared_distances=[]
sscore=[]
chscore=[]
dbscore=[]

ks=np.arange(2,20)
for k in ks:
    brc=Birch(n_clusters=k).fit(X)
    sscore.append(silhouette_score(X, brc.labels_))
    chscore.append(calinski_harabasz_score(X, brc.labels_))
    dbscore.append(davies_bouldin_score(X, brc.labels_))
    

fig, axs=plt.subplots((3), figsize=(15,15))

axs[0].plot(ks, sscore)
axs[0].set_xlabel('Number of Clusters')
axs[0].set_ylabel('Score (greater is better)')
axs[0].set_title('Silhouette Coefficient')
axs[0].set_xticks(ks)

axs[1].plot(ks, chscore)
axs[1].set_xlabel('Number of Clusters')
axs[1].set_ylabel('Score (greater is better)')
axs[1].set_title('Calinski Harabasz Index')
axs[1].set_xticks(ks)

axs[2].plot(ks, dbscore)
axs[2].set_xlabel('Number of Clusters')
axs[2].set_ylabel('Score (lower is better)')
axs[2].set_title('Davies Bouldin Index')
axs[2].set_xticks(ks)
plt.show()
```

VALIDATION

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.model_selection import KFold
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3, init= 'k-means++')
kf=KFold(n_splits=5)
score=cross_val_score(model,X,y,cv=kf, scoring='accuracy')
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps = 0.82, min_samples = 50)
kf=KFold(n_splits=5)
score=cross_val_score(model,X,y,cv=kf, scoring='accuracy')
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

```python
from sklearn.cluster import OPTICS, cluster_optics_dbscan
model = OPTICS(max_eps=0.82, min_samples=50, cluster_method='xi', metric='minkowski')
kf=KFold(n_splits=5)
score=cross_val_score(model,X,y,cv=kf, scoring='accuracy')
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

```python
from sklearn.cluster import Birch
model = Birch(branching_factor = 50, n_clusters= 3, threshold = 1.5)
kf=KFold(n_splits=5)
score=cross_val_score(model,X,y,cv=kf, scoring='accuracy')
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

The algorithm with the highest V-measure Score, Adjusted Rand Index, Adjusted Mutual Information, Silhouette Coefficient, Calinski Harabasz Index and the lowest Davies Bouldin Index is the BIRCH algorithm.
